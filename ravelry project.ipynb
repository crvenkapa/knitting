# Analysis of Ravelry.com

This is a brief write up of some basic data analysis we have done with data scraped from ravelry.com, a knitting and crocheting community. Our goal is to ultimately build a recommender system which will suggest possible knitting/crocheting patterns to users based on past projects they have done and the rating they have given them. For now, we do not have sufficient data to complete this project but hope to get the official API within the next couple of weeks with which we will be able to do more. 

```python
import os
import pandas as pd

d = {}
cat_list = []
for f in os.listdir("data"):
	if "ext" in f:
		cat_name = f.split('_')[0]
		d[cat_name] = pd.read_csv(os.path.join("data", f))
        
df = pd.concat(d)
```

The data frame imported in the above code consists of about 5,500 patterns from about 50 categories. For each pattern, we have the number of users who are working on the pattern or who have completed it. A subset of these users have also given the pattern a star rating (1 to 5) and a difficulty rating (1 to 10), which is stored as both a discrete list and an averaged value. Finally, we also have the number of users who have the pattern in their queue of projects to do. 

Below is a visualization of the top 10 categories according to the number of projects completed or in the process of being completed. Each category is limited to the 100 most popular patterns. 

```python
cats = df.nprojects.groupby(level=0).sum()
cats.sort(ascending=False)
cats[:10].plot(kind='bar')
```

The following code builds a histogram of the total number of votes for each rating level for both the star and difficulty ratings. As we can see, the most common star rating is '5 stars' and the most common difficulty ratings are '1' to '3', ie. most patterns users choose to work on are liked and considered to be easy. This isn't surprising considering that some patterns may take several months to complete, implying that a user is unwilling to spend time working on something that they do not enjoy. 

```python
# Star votes:

starhist = zeros(5)

for e in df.star_rating:
	votes = array([int(s) for s in e.split(';')])
	starhist = starhist + votes
    
bar(arange(1,6), starhist, align='center')
xlabel('Star Rating')
ylabel('Number of Votes')
```

```python
# Difficulty votes:

diffhist = zeros(10)

for e in df.diff_rating:
	votes = array([int(s) for s in e.split(';')])
	diffhist = diffhist + votes
    
bar(arange(1,11), diffhist, align='center')
xlabel('Difficulty Rating')
ylabel('Number of Votes')
```

We now analyze the data in more detail. The following two sets of scatter plots analyze the relation between the number of projects completed to other variables and the number of projects queued by users to other variables. 

The following three scatter plots compare the number of projects completed or currently being worked on to the age of the particular pattern in days, the average difficulty rating and the average star rating for the pattern. 

```python
fig = figure(figsize=(15,8))
subplot(1,3,1)
scatter(df.project_age, df.nprojects)
axis([-100,2800,-100,25000])
ylabel('Number of projects for a particular pattern')
xlabel('Pattern age')
subplot(1,3,2)
scatter(df.avgdiff, df.nprojects)
axis([0.5,10.5,-100,25000])
xlabel('Average difficulty rating')
subplot(1,3,3)
scatter(df.avgstars, df.nprojects)
axis([0.5,5.5,-100,25000])
xlabel('Average star rating')
```

The following four scatter plots compare the number of projects queued by users to the age of the particular pattern in days, the number of projects for the pattern, the average difficulty rating and the average star rating for the pattern. 

```python
fig = figure(figsize=(15,8))
subplot(1,4,1)
scatter(df.project_age, df.nqueues)
axis([-100,2800,-100,15000])
ylabel('Number of projects queued for a particular pattern')
xlabel('Pattern age')
subplot(1,4,2)
scatter(df.nprojects, df.nqueues)
axis([-100,25000,-100,15000])
xlabel('Number of projects')
subplot(1,4,3)
scatter(df.avgdiff, df.nqueues)
axis([0.5,10.5,-100,15000])
xlabel('Average difficulty rating')
subplot(1,4,4)
scatter(df.avgstars, df.nqueues)
axis([0.5,5.5,-100,15000])
xlabel('Average star rating')
```

### Conclusions

It is hard to make any conclusions of the above data, but we can observe that 

* age of a pariculr pattern has an obvious relationship to the number of projects completed or the number queued by users for that pattern;
* the majority of projects completed fall in the easy difficulty rating and have a high star rating, which can be explained by the fact that these projects may take a long time to complete and are therefore not trivial commitments;
* the number of projects queued is correlated to the number of projects completed or in the process of being completed;
* and, similarly, the number of projects queued for a paricular pattern increases with an easier difficulty rating and a higher star rating.
